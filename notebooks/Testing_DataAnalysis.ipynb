```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import lvpyio as lv
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader

data_dir = r"C:\Users\s174006\OneDrive - Danmarks Tekniske Universitet\Deep Learning Turbulence Project\ML_project_data\100img\ImgPreproc(no_subtract)"
save_dir = r"C:\Users\s174006\OneDrive - Danmarks Tekniske Universitet\Deep Learning Turbulence Project"
```

```python
#buffer = lv.read_buffer(path+'/'+"B00001.im7")
frame_number = 1
def load_frame(frame_number, camera_number=0):
    buffer = lv.read_buffer(data_dir +'/' + f"B{str(frame_number+1).zfill(5)}.im7")
    image = buffer[camera_number].as_masked_array().data
    return image
def Threshholding(dataMatrix, a,b):
    return(np.clip(dataMatrix, a, b)/(b-a))
dataMatrix= load_frame(frame_number)
dataMatrix_cliped = Threshholding(dataMatrix, 0, 30)
plt.imshow(dataMatrix_cliped)
plt.show()
```
